```python
#Speeding offenses
SpeedingTickets = df.loc[df['SPEED_IND'] == "Y"]
SpeedingTickets
```

```python
#Count of speeding fine in 2018, out of a total of 49040 offences
SpeedingTickets['OFFENCE_DESC'].count()
```

```python
#Bar chart of Types of speeding
df10 = pd.DataFrame
ax = SpeedingType.plot.bar(x='Type_of_Ticket', y='Amount_of_Offences', rot=0, figsize=(20, 10))
```
